```python
locations = []
for event_mention in event_mentions.findall('event_mention'):
    #print (event_mention.attrib['ID'])
    for event_mention_argument in event_mention.findall('event_mention_argument'):
        if (event_mention_argument.attrib['ROLE'] == 'Location'):
            #print(event_mention_argument[0].text)
            locations.append((event_mention_argument[0].text, 
                              int(event_mention_argument[0].attrib['START']), 
                              int(event_mention_argument[0].attrib['END'])))
            
print(locations)            
```

# the main program

```python
def extract_locations_xml(root):
    event_mentions = root[0]
    locations = []
    for event_mention in event_mentions.findall('event_mention'):
        #print (event_mention.attrib['ID'])
        for event_mention_argument in event_mention.findall('event_mention_argument'):
            if (event_mention_argument.attrib['ROLE'] == 'Location'):
                #print(event_mention_argument[0].text)
                locations.append((event_mention_argument[0].text, 
                              int(event_mention_argument[0].attrib['START']), 
                              int(event_mention_argument[0].attrib['END'])))

    return (locations)            
```

```python
from itertools import zip_longest
import glob

import pandas as pd
import numpy as np
```

```python
location = '../Makarov/versions_combined/all/'
file_names = glob.glob(location+'*.xml')
print(file_names)
```

```python
try:
    tree_files = [ET.parse(file_name) for file_name in file_names]
except OSERROR:
    print('Error in parsing the xml files')
```

```python
roots = [tree.getroot() for tree in tree_files]
roots_ID = [root.attrib['DOCID'] for root in roots]
roots_ID
```

```python
places = [extract_locations_xml(root) for root in roots]
```

```python
df = pd.DataFrame()
df['DOCID'] = np.asarray(roots_ID)
df['Locations'] = np.asarray(places)
df
```
