```python
def cut_txt(text):
    return text[0:len(text)-4]
```

```python
extracted_df = pd.read_pickle("./makarov_documents_extracted_df.pkl")
#extracted_df['file_name']  = extracted_df['file_name'].apply(cut_txt)
extracted_df
```

# Article level measurment

```python
correct = 0
for index, row in locations_df.iterrows():
    X = row['Locations']
    #print(type(X))
    #print(X)
    DOCID = row['DOCID']
    
    list_dic = extracted_df.loc[ extracted_df['file_name'] == DOCID, 'locations_spacy+nominatim']
    #print(list_dic.iloc[0])
    #print(type(list_dic.iloc[0]))
    Y = [(dic.get('original_text'), dic.get('start_char'), dic.get('end_char')) for dic in list_dic.iloc[0]]
    
    if set(X).issubset(set(Y)):
        correct+=1
        print('CORRECT   NO:', index, '  DOCID:', DOCID, '=>')
    else:
        print('WRONG   NO:', index,'  DOCID:', DOCID, '=>')

    print(set(X))
    print('')
    print(set(Y))
    print('-----------------------------------------------------------------------------------------------------')
```

```python

print('article level Corrects:', correct)
```

# place level

```python
correct = 0
totall = 0
for index, row in locations_df.iterrows():
    X = row['Locations']
    totall+=len(X)
    DOCID = row['DOCID']
    
    list_dic = extracted_df.loc[ extracted_df['file_name'] == DOCID, 'locations_spacy+nominatim']
    Y = [(dic.get('original_text'), dic.get('start_char'), dic.get('end_char')) for dic in list_dic.iloc[0]]
    
    
    for place in set(X):
        if place in set(Y):
            correct+=1
            print('CORRECT   NO:', index, '  DOCID:', DOCID, '=>  ', place)
        else:
            print('WRONG     NO:', index, '  DOCID:', DOCID, '=>  ', place)
            print(Y)
            print('')
```

```python
print('place level Corrects:', correct)
print('place level totall:', totall)
```
