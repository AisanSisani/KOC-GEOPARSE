```python
import spacy
from geopy import geocoders
import pandas as pd
```

```python
nominatimGeo = geocoders.Nominatim(user_agent="aisan")
nlp = spacy.load("xx_ent_wiki_sm") #multilingial
```

```python
#location = '../Makarov/'
df = pd.read_csv('makarov_documents.csv')
df
```

```python
i=0
def extract_location(text):
    global i
    try:
        doc = nlp(text)
        locations = []
        for ent in doc.ents:
            if ent.label_== 'LOC':
                loc = nominatimGeo.geocode(ent.text, timeout= 100)

                dictionary = {'original_text':ent.text,
                              'start_char':ent.start_char,
                              'end_char':ent.end_char,
                              'label':ent.label_,
                              'found_text': loc,
                              'latitude': loc.latitude if (loc != None) else '-',
                              'longitude': loc.longitude if (loc != None) else '-'
                             }
                locations.append(dictionary)
        return locations
    except:
        print('Error')
        return None
    finally:
        print(i)
        i+=1
        
```

```python
df['locations_spacy+nominatim'] = df['text'].apply(extract_location)
```

```python
df.loc[17,'locations_spacy+nominatim'] = extract_location(df.loc[17,'text'])
df.loc[34,'locations_spacy+nominatim'] = extract_location(df.loc[34,'text'])
df.loc[88,'locations_spacy+nominatim'] = extract_location(df.loc[88,'text'])
df.loc[118,'locations_spacy+nominatim'] = extract_location(df.loc[118,'text'])
df.loc[210,'locations_spacy+nominatim'] = extract_location(df.loc[210,'text'])
df.loc[220,'locations_spacy+nominatim'] = extract_location(df.loc[220,'text'])
df.loc[228,'locations_spacy+nominatim'] = extract_location(df.loc[228,'text'])
```

```python
df.to_csv('makarov_documents_extracted.csv',index=False)
df.to_pickle("./makarov_documents_extracted_df.pkl")
```
